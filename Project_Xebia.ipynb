```python
def add_title(x):

    if "Mr." in x or "Don." in x:
        return "Mr."
    elif 'Ms.' in x or 'Miss.' in x:
        return "Miss."
    elif "Mrs." in x:
        return "Mrs."
    elif 'Master.' in x:
        return "Master."
    elif "Rev." in x:
        return "Rev."
    else:
        return ""
```

```python
def drop_outliers(df, x):
    first_quartile = df[x].quantile(0.25)
    third_quartile = df[x].quantile(0.75)
    iqr = third_quartile - first_quartile

    low_bound = first_quartile - iqr*1.5
    up_bound = third_quartile + iqr*1.5

    return df[(df[x] >= low_bound) & (df[x] <= up_bound)]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant


def compute_vif(considered_features):

    X = df[considered_features].copy()
    X['intercept'] = 1

    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
```

```python
considered_features = ['Pclass', 'SibSp', 'Fare', 'Sex_male', 'title_Mr.', 'title_Miss.', 'title_Mrs.']

compute_vif(considered_features).sort_values('VIF', ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Survived']
X = df.drop('Survived', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=17)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split

forest = RandomForestClassifier(random_state=17)
tree = DecisionTreeClassifier(random_state=17)


# Set parameters for hyperparameter tuning
forest_params = {
               'n_estimators' : [50,100, 200, 400],
               'criterion' : ['gini', 'entropy'],
               'max_depth' : [2,3,4,5,6,7,8,9,10],
                 'min_samples_split' : [2,5,10,15,20,25],
                 'min_samples_leaf' : [2,4,6,8,10]
             }


tree_params = {
                 'criterion' : ['gini', 'entropy'],
                 'max_depth' : [2,3,4,5,6,7,8,9,10],
                 'min_samples_split' : [2,5,10,15,20,25],
                 'min_samples_leaf' : [2,4,6,8,10]
              }

scoring = 'accuracy'
```

```python
def select_best_model(algorithm1=tree, algorithm2=forest, params1=tree_params, params2=forest_params, X = X_train, y=y_train):

    tree_rscv = RandomizedSearchCV(tree, tree_params, cv=8, scoring=scoring, random_state=17)
    tree_rscv.fit(X_train, y_train)

    forest_rscv = RandomizedSearchCV(forest, forest_params, cv=8, scoring=scoring, random_state=17)
    forest_rscv.fit(X_train, y_train)

    tree_score = tree_rscv.best_score_
    forest_score = forest_rscv.best_score_

    if tree_score > forest_score:
        best_model = tree_rscv.best_estimator_
    else:
        best_model = forest_rscv.best_estimator_

    tree_score = 'Decision Tree best accuracy score: ' + str(tree_score)
    forest_score = 'Random Forest best accuracy score: ' + str(forest_score)


    return (tree_score, forest_score,best_model)
```

```python
tree_score, forest_score, best_model = select_best_model()
print(tree_score, "\n", forest_score)
```
